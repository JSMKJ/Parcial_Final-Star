El objetivo es predecir el precio del automÃƒÆ’Ã‚Â³vil. Para mÃƒÆ’Ã‚Â¡s detalles puede visitar el siguiente enlace: [datos](https://www.kaggle.com/jpayne/852k-used-car-listings).

```python
#librerias
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# ImportaciÃƒÆ’Ã‚Â³n de librerÃƒÆ’Ã‚Â­as
%matplotlib inline
import pandas as pd

# Lectura de la informaciÃƒÆ’Ã‚Â³n de archivo .csv
data = pd.read_csv('https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/datasets/dataTrain_carListings.zip')

# Preprocesamiento de datos para el taller
data = data.loc[data['Model'].str.contains('Camry')].drop(['Make', 'State'], axis=1)
data = data.join(pd.get_dummies(data['Model'], prefix='M'))
data = data.drop(['Model'], axis=1)

# VisualizaciÃƒÆ’Ã‚Â³n dataset
data.head()
```

```python
# SeparaciÃƒÆ’Ã‚Â³n de variables predictoras (X) y variable de interÃƒÆ’Ã‚Â©s (y)
y = data['Price']
X = data.drop(['Price'], axis=1)
```

```python
# SeparaciÃƒÆ’Ã‚Â³n de datos en set de entrenamiento y test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Punto 1 - Random forest con librerÃƒÆ’Ã‚Â­a

En la celda 1, usando la librerÃƒÆ’Ã‚Â­a sklearn entrenen un modelo de Randon Forest para clasificaciÃƒÆ’Ã‚Â³n  y comenten sobre el desempeÃƒÆ’Ã‚Â±o del modelo.

```python
# Celda 1

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Asegurar que existen X_train/X_test/y_train/y_test
try:
    X_train
    X_test
    y_train
    y_test
except NameError:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Modelo base (regresion)
rf = RandomForestRegressor(random_state=1, n_estimators=100, n_jobs=-1)
rf.fit(X_train, y_train)

# Desempeno en test
y_pred = rf.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print(f"Test MAE: {mae:,.2f}")
print(f"Test RMSE: {rmse:,.2f}")
print(f"Test R2: {r2:,.3f}")

# Validacion cruzada (MAE)
cv_mae = -cross_val_score(rf, X, y, cv=5, scoring="neg_mean_absolute_error")
print(f"5-fold CV MAE (mean+/-std): {cv_mae.mean():,.2f} +/- {cv_mae.std():,.2f}")
```

### Punto 2 - CalibraciÃƒÆ’Ã‚Â³n de parÃƒÆ’Ã‚Â¡metros Random forest

En la celda 2, calibren los parÃƒÆ’Ã‚Â¡metros max_depth, max_features y n_estimators del modelo de Randon Forest para clasificaciÃƒÆ’Ã‚Â³n, comenten sobre el desempeÃƒÆ’Ã‚Â±o del modelo y describan cÃƒÆ’Ã‚Â³mo cada parÃƒÆ’Ã‚Â¡metro afecta el desempeÃƒÆ’Ã‚Â±o del modelo.

```python
# Celda 2

from sklearn.model_selection import cross_val_score



# Barrido de n_estimators (como en el tutorial)

estimator_range = range(50, 401, 50)

mae_scores = []

for n in estimator_range:

    rf_tmp = RandomForestRegressor(n_estimators=n, random_state=1, n_jobs=-1)

    mae_cv = -cross_val_score(rf_tmp, X, y, cv=5, scoring='neg_mean_absolute_error').mean()

    mae_scores.append(mae_cv)

plt.figure()

plt.plot(list(estimator_range), mae_scores, marker="o")

plt.xlabel('n_estimators')

plt.ylabel('CV MAE (lower is better)')

plt.title('Tuning n_estimators')



# Barrido de max_features

feature_range = list(range(1, X.shape[1] + 1, max(1, X.shape[1]//10)))

mae_scores2 = []

for m in feature_range:

    rf_tmp = RandomForestRegressor(max_features=m, random_state=1, n_jobs=-1)

    mae_cv = -cross_val_score(rf_tmp, X, y, cv=5, scoring='neg_mean_absolute_error').mean()

    mae_scores2.append(mae_cv)

plt.figure()

plt.plot(feature_range, mae_scores2, marker="o")

plt.xlabel('max_features')

plt.ylabel('CV MAE (lower is better)')

plt.title('Tuning max_features')
```

```python
#Encontrar los mejores hiperparametros: 
from sklearn.model_selection import GridSearchCV
\n
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['sqrt', 'log2', None, max(1, X.shape[1]//3), max(1, X.shape[1]//2)],
    'max_depth': [None, 10, 20, 30]
}
rf_gs = RandomForestRegressor(random_state=1, n_jobs=-1)
grid = GridSearchCV(rf_gs, param_grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
grid.fit(X, y)
print('Best params:', grid.best_params_)
print('Best CV MAE:', -grid.best_score_)
best_params = grid.best_params_
\n
# Entrenar el modelo con los mejores hiperparÃƒÂ¡metros encontrados
rf_best = RandomForestRegressor(random_state=1, n_jobs=-1, **best_params)
rf_best.fit(X_train, y_train)
y_pred_best = rf_best.predict(X_test)
print(f'Test MAE: {mean_absolute_error(y_test, y_pred_best):,.2f}')
print(f'Test RMSE: {mean_squared_error(y_test, y_pred_best, squared=False):,.2f}')
print(f'Test R2: {r2_score(y_test, y_pred_best):,.3f}')
```

```python
#Importancia de los predictores

import pandas as pd

importances = pd.DataFrame({

    'feature': X.columns,

    'importance': rf_best.feature_importances_

}).sort_values('importance', ascending=False)

importances.head(20)
```
